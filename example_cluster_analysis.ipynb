```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


#Cluster using euclidean distance via kmeans as a comparison

df_continuous = df

# Initialize the scaler
scaler = StandardScaler()

# Scale the continuous data
df_continuous_scaled = scaler.fit_transform(df_continuous)

# Specify the number of clusters (adjust as needed)
num_clusters = 4

# Initialize the KMeans model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the model to the data without NaN values
kmeans.fit(df_continuous)

# Add the cluster labels to the original DataFrame
df['Cluster'] = pd.Series(index=df.index, data=-1)  # Initialize with -1 for rows with NaN
df.loc[df_continuous.index, 'Cluster'] = kmeans.labels_

# Reduce dimensionality using t-SNE
tsne = TSNE(n_components=2, random_state=42)
df_tsne = pd.DataFrame(tsne.fit_transform(df_continuous), columns=['TSNE1', 'TSNE2'])

# Plot the clusters in 2D
plt.figure(figsize=(10, 8))
scatter = plt.scatter(df_tsne['TSNE1'], df_tsne['TSNE2'], c=df.loc[df_continuous.index, 'Cluster'], cmap='viridis')
plt.title('t-SNE Visualization of Clusters')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(*scatter.legend_elements(), title='Clusters')
plt.show()
```
